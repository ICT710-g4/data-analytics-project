# To DataFrame type

```python
def load_data(file_name):
    file = open(file_name, "r")
    file_string = file.read()
    file_string_list = file_string.split("\x00\n")
    for i in range(len(file_string_list)):
        file_string_list[i] = file_string_list[i].split(",").copy()
    return file_string_list[:-1]
```

```python
def convert2df(string_list, col_name=None):
    df = pd.DataFrame(string_list)
    if col_name is not None:
        df.columns = col_name
    df = df.astype(int)
    return df
```

```python
col_name = ("roll", "pitch", "yaw", "acc_x", "acc_y", "acc_z", "label")
```
